```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Superstore.csv")
```

```python
# - Summary of the data
df.info()
```

```python
# - Shape of the data

df.shape
```

```python
# - Statistical Summary of all columns (all datatypes)
df.dtypes
```

```python
# - Missing Value Detection
df.isna().sum()
```

```python
# - Drop Missing Values
df.duplicated().sum()
```

```python
# - Drop Missing Values
df.drop_duplicates()
```

```python
# - Drop RowId column
df.drop(columns=["RowId"] , axis=1)
```

```python
# - Convert date columns into a datetime format
df["Ship Date"]=pd.to_datetime(df["Ship Date"],errors="coerce")
df["Order Date"]=pd.to_datetime(df["Order Date"],errors="coerce")
```

```python
# - Find the correlation between all numeric columns
data_frame=df
data_frame=data_frame.drop(columns=["Order ID","Customer ID",
                                    "Customer Name","Segment",
                                    "Country/Region","City",
                                    "Region","Product ID",
                                    "Category","Sub-Category","Product Name","Ship Mode","State","RowId"],axis=1)
```

```python
data_frame.corr()
```

```python
# 2. How many unique customers are there, and which customer has placed the most orders? (2 marks)
df["Customer Name"].nunique()
```

```python
data=df.groupby("Customer Name")["Order ID"].count().reset_index()
data.max()["Customer Name"]
```

```python
# 3. How does the average order quantity vary across different categories? (2 marks)
df.groupby("Category")["Quantity"].mean().reset_index()
```

```python
# B. Visualization:
# 5. What are the top 5 most frequently ordered products? (4 marks)
top5=df["Product Name"].value_counts().head(5).reset_index()
top5.columns = ["Product Name", "Count"] 
sns.barplot(x="Product Name",y="Count",data=top5)
plt.title("Top 5 Most Frequently Ordered Products")
plt.xlabel("Number of Orders")
plt.ylabel("Product Name")
plt.xticks(rotation=75)
plt.show()
```

```python
# 6. What is the overall profit performance of each sub-category? (4 marks) 
sub=df.groupby("Sub-Category")["Profit"].sum().reset_index()
sns.barplot(x="Sub-Category",y="Profit",data=sub)
plt.title("Overall Profit Performance of Each Sub-Category")
plt.xlabel("Total Profit")
plt.ylabel("Sub-Category")
plt.xticks(rotation=75)
plt.show()
```

```python
# 7. What is the distribution of sales across different states? (4 marks)
sale_region=df.groupby("Region")["Sales"].sum().reset_index()
sale_region
sns.barplot(x="Region",y="Sales",data=sale_region)
plt.title("Distribution of Sales Across Different States",)
plt.xlabel("Total Sales")
plt.ylabel("State")
plt.show()
```

```python
# 8. What is the distribution of customers across segments (e.g., Consumer, Corporate)? (4 marks)
sg_data=df.groupby("Segment")["Customer ID"].nunique().reset_index()
sg_data.columns=["Segment","Customer Count"]
sns.barplot(x="Segment",y="Customer Count",data=sg_data)
plt.title("Distribution of Customers Across Segments")
plt.xlabel("Segment")
plt.ylabel("Number of Unique Customers")
plt.show()
```

```python
top_categories=df.groupby("Category")["Sales"].sum().reset_index()
top_categories=top_categories.sort_values(by="Sales",ascending=False)
sns.barplot(x="Category", y="Sales", data=top_categories)
plt.title("Top-Selling Categories")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
top_Subcategories=df.groupby("Sub-Category")["Sales"].sum().reset_index()
top_Subcategories=top_Subcategories.sort_values(by="Sales",ascending=False)
sns.barplot(x="Sub-Category", y="Sales", data=top_Subcategories)
plt.title("Top-Selling Sub-Categories")
plt.xlabel("Sub-Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=75)
plt.show()
```

```python
# 10. How do monthly or yearly sales trends look over time?  (10 marks)
#     (Hint: study and use df.resample() for monthly data)

df.set_index('Order Date', inplace=True)  
yearly_sales = df['Sales'].resample('Y').sum().reset_index()
sns.barplot(x=yearly_sales['Order Date'].dt.year, y='Sales', data=yearly_sales)
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()
```

```python
  
monthly_sales = df['Sales'].resample('M').sum().reset_index()
sns.barplot(x=monthly_sales['Order Date'].dt.year, y='Sales', data=monthly_sales)
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```
